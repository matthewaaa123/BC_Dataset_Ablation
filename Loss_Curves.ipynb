 # Plot Loss Curves for Each Dataset Size

 ---

 The following code plots a loss curve that shows the training process for each dataset size.

```python
import pandas as pd
import matplotlib.pyplot as plt

loss_full = pd.read_csv('loss_full.csv', header=None)
loss_half = pd.read_csv('loss_half.csv', header=None)
loss_quarter = pd.read_csv('loss_quarter.csv', header=None)
loss_tenth = pd.read_csv('loss_tenth.csv', header=None)
loss_twentieth = pd.read_csv('loss_twentieth.csv', header=None)
loss_hundredth = pd.read_csv('loss_hundredth.csv', header=None)

plt.figure(figsize=(12, 7))

plt.plot(loss_full[0], loss_full[2], label='99,522 Episodes', linewidth=2, color='#1f77b4')
plt.plot(loss_half[0], loss_half[2], label='50,000 Episodes', linewidth=2, color='#ff7f0e')
plt.plot(loss_quarter[0], loss_quarter[2], label='25,000 Episodes', linewidth=2, color='#2ca02c')
plt.plot(loss_tenth[0], loss_tenth[2], label='10,000 Episodes', linewidth=2, color='#d62728')
plt.plot(loss_twentieth[0], loss_twentieth[2], label='5,000 Episodes', linewidth=2, color='#9467bd')
plt.plot(loss_hundredth[0], loss_hundredth[2], label='1,000 Episodes', linewidth=2, color='#8c564b')

plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Loss', fontsize=16)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.legend(fontsize=14)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
